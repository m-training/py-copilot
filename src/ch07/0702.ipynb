指定した単語でGoogleで検索するコードを書いて。


```python
import webbrowser

def google_search(keyword):
    search_url = f"https://www.google.com/search?q={keyword}"
    webbrowser.open(search_url)

# Example usage
google_search("GitHub Copilot")
```

```python
import requests

def search_word(keyword):
    search_url = f"https://www.google.com/search?q={keyword}"
    response = requests.get(search_url)
    if response.status_code == 200:
        return response.text
    else:
        return None

# Example usage
search_results = search_word("GitHub Copilot")
print(search_results)
```

```python
from bs4 import BeautifulSoup

import requests

def search_word(keyword):
    search_url = f"https://www.google.com/search?q={keyword}"
    response = requests.get(search_url)
    if response.status_code == 200:
        return response.text
    else:
        return None

# Example usage
search_results = search_word("GitHub Copilot")

# Convert the search_results string to HTML format
html_doc = search_results

# Create a BeautifulSoup object
soup = BeautifulSoup(html_doc, 'html.parser')

# Find all H3 tags
h3_tags = soup.find_all('h3')

# Print the H3 tags
for h3 in h3_tags:
    print(h3.text)
```

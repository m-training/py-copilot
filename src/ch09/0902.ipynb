```python
import matplotlib.pyplot as plt

# Data for the bar graph
x = ['A', 'B', 'C', 'D']
y = [10, 5, 12, 8]

# Create the bar graph
plt.bar(x, y)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph')

# Display the graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

# Data for the bar graph
x = ['A', 'B', 'C', 'D', 'E']
y = [random.randint(1, 100) for _ in range(len(x))]

# Create the bar graph
plt.bar(x, y)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph')

# Display the graph
plt.show()
```
